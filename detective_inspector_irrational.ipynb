```python
import math
import cmath
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#Setup graph
mpl.rcParams['font.size'] = 20
plt.xlim(-1, 1)
plt.ylim(-1,1)
plt.axhline()
plt.axvline()
plt.xlabel("Real")
plt.ylabel("Imaginary")

#---Task 1---
#--1.1--

z_num = (1+1j)
z_den = math.sqrt(2) * (1j**4) * ((1+1j)/(abs((1+1j))))**40
z_initial = z_num/z_den
u = math.sqrt(2)/(1+1j)

plt.plot(z_initial.real, z_initial.imag, 'o') #Plot Initial Point

total = 0
z_position = z_initial

#Movements
for i in range(64):
    z_position /= u
    total += z_position
    plt.plot(z_position.real, z_position.imag, 'o', c ='r') #Plots moves

mean = total/64
plt.plot(mean.real, mean.imag, 'o', c ='g') #Key card coordinates
plt.text(0.05,0.05,"Key Card")

#Answer = The keycard is located in coordinates (0,0)

#--1.2--
```

```python
import math
import cmath
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#Setup graph
mpl.rcParams['font.size'] = 15
plt.xlim(-2, 2)
plt.ylim(-2,2)
plt.axhline()
plt.axvline()
plt.xlabel("Real")
plt.ylabel("Imaginary")

#---Task 2---
#--2.1--

z_num = math.sqrt(2)*(1+1j)
z_den = ((1+1j)/(math.sqrt(2)))**40
z_initial = z_num/z_den
u = math.sqrt(2)/(1+1j)

plt.plot(z_initial.real, z_initial.imag, 'o', c = 'b') #Initial Point


z_position = z_initial

#Movements
moves = 0
total = 0
while(True):
    z_position /= u
    plt.plot(z_position.real, z_position.imag, 'o', c = 'r') #Plots moves
    moves += 1
    total += z_position
    mean = total/moves
    if abs(mean.real*10).__trunc__() == 0 and abs(mean.imag*10).__trunc__() == 0:
        print(mean.real, mean.imag)
        plt.plot(z_initial.real, z_initial.imag, 'o', c = 'b') #Clarify Initial Point
        plt.text(z_initial.real-1, z_initial.imag,"Initial Point")
        plt.plot(mean.real, mean.imag, 'o', c = 'g') #Keycard
        plt.text(0.1,0.1,"Key Card")
        break

print(moves)
#Answer = The minimum number of movements is 8

#--2.2--
```

```python
import math
import cmath
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#Setup graph
mpl.rcParams['font.size'] = 12
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(0,100)

#---Task 4---
#--4.1--

x_initial = -3*40
x_final = 3*40
x = np.arange(x_initial, x_final, 0.001)
y_num = (x**3) - (2*40*(x**2)) + ((40**2)*x)
y_den = (x-2*40)**2
y = np.sqrt(y_num/y_den)
plt.plot(x,y, 'r--')
plt.legend([r'$\sqrt{\frac{x^3 - 80x^2 + 1600x}{(x-80)^2}}$'])

#Programatically getting the x-value that makes the function head towards +inf
x = np.arange(0, x_final, 0.001)
y_num = (x**3) - (2*40*(x**2)) + ((40**2)*x)
y_den = (x-2*40)**2
y = np.sqrt(y_num/y_den)
for coord in zip(x,y):
    if coord[1] == np.inf:
        print(f"y({coord[0]}) = +inf")

#To figure out the x-value that makes the function head towards +inf, we make the denominator equal to 0:
# (x-80)^2 = 0
# (x-80)(x-80)=0
# x = 80

#Answer = Safe positions: x ∈ [0, 80) ∪ (80, +∞) OR {x ∈ R | x ≥ 0 and x ≠ 80}

#--4.2--

# Set the expression inside the square root to be non-negative i.e. (x^3-80x^2+1600x)/(x-80)^2 ≥ 0, we know that the safe positions are only going to consist of positive values
# Solve the numerator first so (x^3-80x^2+1600x) = 0. We get that x = 0, x = 40 since x(x-40)^2 is the same as (x^3-80x^2+1600x)
# Then the denominator (x-80)^2 = 0, so we get that x is 80
# The numerator indicates that the graph starts at 0 and intersects the x-axis at 40
# The denominator indicates that at 80, a divide by 0 occurs; the graph heads towards +inf, therefore not being a safe position
```
